```python
# caregiver class
class Caregiver:
    def __init__(self, name, weekly_hours):
        self.name = name
        self.weekly_hours = weekly_hours
        self.rate = 20  # hourly rate in dollars

    def weekly_pay(self):
        # weekly pay

        return self.weekly_hours * self.rate


# generate pay report
def generate_pay_report(caregivers):
    total_weekly_pay = 0
    total_monthly_pay = 0
    weekly_report = []

    for caregiver in caregivers:
        weekly_pay = caregiver.weekly_pay()
        total_weekly_pay += weekly_pay #Mutiplies hours by rate (20) 
        total_monthly_pay += weekly_pay * 4  # 4 weeks per month

# caregivers 
caregivers = []

# How many caregivers will be on the pay report 
num_caregivers = int(input("How many caregivers would you like to input? "))

for i in range(num_caregivers):
    name = input(f"Enter the name of caregiver {i+1}: ") #Name of each caregiver, {i+1} allows to see which number caregiver you are on 
    weekly_hours = int(input(f"How many hours did {name} work this week? "))  # input to integer
    caregivers.append(Caregiver(name, weekly_hours)) #Adds the values to the caregivers list 

def generate_html_report(caregivers): #Function to make HTML report 
    total_weekly_pay = 0
    total_monthly_pay = 0
    weekly_report = []

    for caregiver in caregivers:
        weekly_pay = caregiver.weekly_pay()
        weekly_report.append((caregiver.name, caregiver.weekly_hours, weekly_pay))#Adds caregiver info to the weekly report 
        total_weekly_pay = weekly_pay 
        total_monthly_pay = weekly_pay * 4  # 4 weeks per month

    #HTML structure 
    report = """
    <html>
    <head><title>Caregiver Pay Report</title></head>
    <body>
    <h2>Weekly Pay Report</h2>
    <table border="1" cellpadding="5">
        <tr>
            <th>Caregiver</th>
            <th>Hours Worked</th>
            <th>Weekly Pay</th>
        </tr>
    """
    
    for caregiver in weekly_report: #Adds the caregiver
        report += f"""
        <tr>
            <td>{caregiver[0]}</td>
            <td>{caregiver[1]}</td>
            <td>${caregiver[2]:.2f}</td>
        </tr>
        """
    
    report += f"""
        <tr>
            <td><b>Total Weekly Pay</b></td>
            <td></td>
            <td><b>${total_weekly_pay:.2f}</b></td> 
        </tr>
        <tr>
            <td><b>Total Monthly Pay (4 weeks)</b></td>
            <td></td>
            <td><b>${total_monthly_pay:.2f}</b></td>
        </tr>
    </table>
    </body>
    </html>
    """
    
    return report


# Generate the HTML report
html_report = generate_html_report(caregivers)

# Save the report to an HTML file
with open("caregiver_pay_report.html", "w") as file:
    file.write(html_report)
    print("HTML successfully made") #System output to show it was successfully
```
